```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# 1. 데이터 로드 및 준비
iris = load_iris()
X = iris.data  # 특징 데이터
y = iris.target  # 라벨 데이터

# 2. 데이터 표준화 (PCA 수행 전에 표준화가 필요)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. PCA 모델 정의 (주성분 개수 2개로 설정)
pca = PCA(n_components=2)

# 4. PCA 적용 (차원 축소)
X_pca = pca.fit_transform(X_scaled)

# 5. 결과 시각화
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.colorbar(label='Species')
plt.title('PCA of Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# 6. 주성분과 설명된 분산 비율 출력
print("Explained variance ratio (PC1, PC2):", pca.explained_variance_ratio_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# 1. 데이터 로드 및 준비
iris = load_iris()
X = iris.data  # 특징 데이터
y = iris.target  # 라벨 데이터

# 2. 데이터 표준화 (PCA 수행 전에 표준화가 필요)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. PCA 모델 정의 (주성분 개수 2개로 설정)
pca = PCA(n_components=3)

# 4. PCA 적용 (차원 축소)
# 5. 결과 시각화 (3D 플롯)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=y, cmap='viridis', edgecolor='k', s=50)
ax.set_title('PCA of Iris Dataset (3 Components)')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
plt.colorbar(scatter, label='Species')
plt.show()

# 6. 주성분과 설명된 분산 비율 출력
print("Explained variance ratio (PC1, PC2, PC3):", pca.explained_variance_ratio_)
```
